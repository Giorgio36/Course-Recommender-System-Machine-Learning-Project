```python
#Calculatin the recomendation score of all courses for all 1000 users
# We first reload all datasets again, and create an empty dictionary to store the results
test_users_df = pd.read_csv(test_users_url)
profile_df = pd.read_csv(profile_genre_url)
course_genres_df = pd.read_csv(course_genre_url)
res_dict = {}
```

```python
# Only keep the score larger than the recommendation threshold
# The threshold can be fine-tuned to adjust the size of generated recommendations
score_threshold = 10.0
```

```python
def generate_recommendation_scores():
  users = []
  courses = []
  scores = []
  for user_id in test_user_ids:
    test_user_profile = profile_df[profile_df['user'] == user_id]
    test_user_vector = test_user_profile.iloc[0,1:].values

    #getting the unknown course ids for the current user id
    enrolled_courses = test_users_df[test_users_df['user'] == user_id]['item'].to_list()
    unknown_courses = all_courses.difference(enrolled_courses)
    unknown_courses_df = course_genres_df[course_genres_df['COURSE_ID'].isin(unknown_courses)]
    unknown_courses_matrix = unknown_courses_df.iloc[:,2:].values
    unknown_courses_ids = unknown_courses_df['COURSE_ID'].values

    #getting recomendation score for each course
    recommendatation_score = []
    for i in range(unknown_courses_matrix.shape[0]):
      result = np.dot(unknown_courses_matrix[i],test_user_vector)
      recommendatation_score.append(result)

    #appending the results into the user, courses, and scores list
    for i in range(0, len(unknown_courses_ids)):
      users.append(user_id)
      courses.append(unknown_courses_ids[i])
      scores.append(recommendatation_score[i])

  res_dict['USER'] = users
  res_dict['COURSE_ID'] = courses
  res_dict['SCORE'] = scores
  res_df = pd.DataFrame(res_dict, columns = ['USER', 'COURSE_ID', 'SCORE']).reset_index()
  res_df = res_df.rename(columns = {'index':'level_1'})
  top_10 = res_df.groupby('USER')['SCORE'].nlargest(10).reset_index()
  top_10_res_df = pd.merge(top_10, res_df, how = 'left', on = ['USER', 'level_1'])
  top_10_res_df = top_10_res_df[['USER', 'COURSE_ID', 'SCORE_x']].rename(columns={'SCORE_x':'SCORE'})

  top_10_res_df.to_csv('/content/drive/MyDrive/Colab Notebooks/IBM/Capstone/Unsupervised Learning based recomender system/recommendation_results.csv', index = False)
  return top_10_res_df
```

```python
recomendations_df = generate_recommendation_scores()
recomendations_df
```

```python
#Getting the top 10 most recomended courses
top_10_courses = recomendations_df.groupby('COURSE_ID').size().sort_values(ascending = False).head(10)
name_top_10 = top_10_courses.index
course_genres_df[course_genres_df['COURSE_ID'].isin(name_top_10)]['TITLE']
```
